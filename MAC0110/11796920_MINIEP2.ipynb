## Parte 1 - Função soma2(x, y)

Cria uma função _soma2_ que recebe os parâmetros _x_ e _y_, e devolve a soma dos dois parâmetros.

```julia
function soma2(x,y)
    soma = x+y 
    return soma
end
```

# Parte 3 - Função calculaDelta(a, b, c)

Cria a função _calculaDelta_ que recebe os parâmetros _a_, _b_ e _c_ e devolve o _delta_.

```julia
function calculaDelta(a,b,c)
    operacao = (b*b + ((-4)*a*c))
    return operacao
end
```
